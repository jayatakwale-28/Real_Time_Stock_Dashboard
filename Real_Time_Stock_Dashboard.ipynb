Import Libraries

```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
```

Download the data

```python
stock = yf.Ticker("AAPL")
df = stock.history(period="6mo")

df.head()
```

```python
#Display the latest price
print("Latest Closing Price")
print(df["Close"].iloc[-1])
```

```python
# Plot the Stock Price
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=df.index,
        y=df["Close"],
        mode="lines",
        name="Closing Price"
    )
)

fig.update_layout(
    title="Apple Stock Price",
    xaxis_title="Date",
    yaxis_title="Price (USD)"
)

fig.show()
```

Add Moving Average

```python
#Calculate 20-days Moving Average
df["MA20"] = df["Close"].rolling(window=20, min_periods=1).mean()

#Calculate 50-days Moving Average
df["MA50"] = df["Close"].rolling(window=50, min_periods=1).mean()
```

Plot Stock price with Moving Average

```python
import plotly.graph_objects as go

fig = go.Figure()

# Closing Price
fig.add_trace(go.Scatter(
    x=df.index,
    y=df["Close"],
    mode="lines",
    name="Closing Price"
))

# Moving Average
fig.add_trace(go.Scatter(
    x=df.index,
    y=df["MA20"],
    mode="lines",
    name="20-Day Moving Average"
))

fig.update_layout(
    title="Apple Stock Price with 20-Day Moving Average",
    xaxis_title="Date",
    yaxis_title="Price (USD)"
)

fig.show()
```

```python
print("Latest Close Price:", df["Close"].iloc[-1])
print("Highest Price:", df["High"].max())
print("Lowest Price:", df["Low"].min())
print("Average Closing Price:", df["Close"].mean())
```

Create the Streamlit App

```python
%%writefile app.py
import streamlit as st
import yfinance as yf
import plotly.graph_objects as go

st.title("📈 Real-Time Stock Market Dashboard")

# Stock selection
stock_symbol = st.selectbox(
    "Select a Stock",
    ["AAPL", "MSFT", "GOOGL", "TSLA", "AMZN", "RELIANCE.NS", "TCS.NS", "INFY.NS"]
)

# Download data
data = yf.Ticker(stock_symbol).history(period="6mo")

# Moving Average
data["MA20"] = data["Close"].rolling(20).mean()

# Stock chart
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=data.index,
    y=data["Close"],
    mode="lines",
    name="Closing Price"
))

fig.add_trace(go.Scatter(
    x=data.index,
    y=data["MA20"],
    mode="lines",
    name="20-Day MA"
))

st.plotly_chart(fig)

# Stock statistics
st.subheader("Stock Statistics")
st.write("Latest Close Price:", round(data["Close"].iloc[-1], 2))
st.write("Highest Price:", round(data["High"].max(), 2))
st.write("Lowest Price:", round(data["Low"].min(), 2))
st.write("Average Closing Price:", round(data["Close"].mean(), 2))
```
